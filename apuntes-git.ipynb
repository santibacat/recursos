# Git básico

Hemos de saber que hay tres estados:
- Working directory: aun no modificados
- Staging area: archivos que ya estan modificados y se van a guardar
- Repositorio

Para empezar, creamos una carpeta que será la carpeta 'local' donde se modifica todo.

En el bash de git, escribimos:

```
git init # Para crear un proyecto nuevo
git status # Para ver el estado (archivos modificados...)
git add archivo.txt # Para añadir un archivo del WD al staging area
```

Otros comandos para añadir o no:

```
git add . # añade todos los archivos sin modificar
git commit # para añadir al repositorio
git commit -m 'Mi primer commit' # Añade al commit dandole un nombre directamente
git log # para ver los commits que se han hecho
```

Si tenemos un archivo sin modificar (en working directory) y queremos volver al archivo original:

```
git checkout -- file.txt # elimina el archivo file.txt y lo restaura del commit
git diff file.txt # muestra las diferencias entre los archivos de ahora y el de antes
```

Si hay archivos que queremos ignorar, creamos un archivo .gitignore en la raiz principal y ahí, en cada linea, añadimos el nombre del archivo o carpeta que no queremos.

```
git add .gitignore
```

Para crear ramificaciones del código:

```
git branch # para saber las ramificaciones
git branch rama2 # para crear una rama llamada rama2
git checkout rama2 # para cambiar a la rama2

# si queremos saber más de ramas buscar cap5 del manual de git que tengo
```

Deshacer cambios y volver a la rama master:

```
git checkout master
git revert HEAD --no-edit # revierte el ultimo comit
```

Para trabajar con codigo de Github:

```
git remote add origin https://github.com/usuario/gitrepositorio.git
  # con esto le decimos donde almacenar el codigo
git push -u origin master
# con esto mandamos el archivo local al remoto

git pull https://github.com/usuario/gitrepositorio
  # Así se rescata entero
```

Taguear versiones:

```
git tag v1 # asigna la version 1 a esta version
git tag # muestra las versiones que hay
git diff v1-v2 # muestra diferencias entre versiones
git tag -d v2 # elimina la etiqueta
```

Mostrar de forma bonita los cambios:

```
git log --pretty=format: '%h %ad | %s%d [%an]' --graph --date=short
  
  # tambien podemos añadir a .gitconfig estas lineas
  # [alias]
  # hist = log --pretty=format: '%h %ad | %s%d [%an]' --graph --date=short
  
  # y así podremos usar 'git log hist'
```

Cambios avanzados:

```
git commit -m 'He hecho un commit y se me ha olvidado una cosa'
git commit --amend -m 'Ahora lo he arreglado y se guarda dentro del mismo commit'
```
